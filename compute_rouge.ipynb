```python
from rouge import RougeScorer
import pandas as pd

df = pd.read_csv('./outputs/inferences/distilkobart-rdrop-demo.tsv', sep='\t')
df.drop('id', axis=1, inplace=True)
df = df.rename_axis('id').reset_index()
df
```

```python
ref_df = df[['id', 'target summary']]
ref_df
```

```python
hyp_df = df[['id', 'predict summary']]
hyp_df
```

```python
rouge_scorer = RougeScorer()
res = rouge_scorer.compute_rouge(ref_df, hyp_df)
```

```python
res
```

```python
print("res['rouge-1']['f']: {:.3f}".format(res['rouge-1']['f']))
print("res['rouge-1']['p']: {:.3f}".format(res['rouge-1']['p']))
print("res['rouge-1']['r']: {:.3f}".format(res['rouge-1']['r']))
print('==========================')
print("res['rouge-2']['f']: {:.3f}".format(res['rouge-2']['f']))
print("res['rouge-2']['p']: {:.3f}".format(res['rouge-2']['p']))
print("res['rouge-2']['r']: {:.3f}".format(res['rouge-2']['r']))
print('==========================')
print("res['rouge-l']['f']: {:.3f}".format(res['rouge-l']['f']))
print("res['rouge-l']['p']: {:.3f}".format(res['rouge-l']['p']))
print("res['rouge-l']['r']: {:.3f}".format(res['rouge-l']['r']))
```
